5. Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn "kg".

```python
weight = patientheartrate['Weight']
for i in range(0, len(weight)):
    x = str(weight[i])
    if "lbs" in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x/2.2)
        y = str(y) + "kgs"
        weight[i] = y
print(patientheartrate)
```

- weight = patientheartrate['Weight']: Lấy cột 'Weight' từ DataFrame và gán nó vào biến 'weight'.

- for i in range(0, len(weight)): Tạo 1 dãy số từ 0 đến độ dài của danh sách 'weight' - 1.

- x = str(weight[i]): Chuyển giá trị trọng lượng tại vị trí 'i' trong danh sách 'weight' thành chuỗi rồi gán vào x.

- if "lbs" in x[-3:]: Kiểm tra chuỗi "lbs" xuất hiện trong ba ký tự cuối cùng của "x". Điều này để kiểm tra xem trọng lượng có được ghi bằng đơn vị pounds hay không.

- x = x[:-3:]: Nếu x chứa "lbs" trong ba ký tự cuối, dòng này loại bỏ ba ký tự cuối cùng khỏi x. Kết quả là chuỗi x chỉ còn lại phần số mà không có "lbs".

- float_x = float(x): Chuyển đổi giá trị chuỗi x (phần số đã loại bỏ "lbs") thành một số thực và gán nó vào biến float_x.

- y = int(float_x / 2.2): Chuyển đổi giá trị từ pounds sang kilograms bằng cách chia float_x cho 2.2 và sau đó chuyển đổi kết quả thành một số nguyên. Kết quả được gán vào biến y.

- y = str(y) + "kgs": Chuyển đổi giá trị số nguyên y thành chuỗi và thêm "kgs" vào cuối chuỗi để biểu thị đơn vị là kilograms. Kết quả được gán vào biến y.

- weight[i] = y: Cập nhật giá trị tại vị trí i trong danh sách weight với giá trị mới là y (trọng lượng đã chuyển đổi sang kilograms).

- patientheartrate_2['PulseRate'].replace({'-' : '0'}, regex=True): 

=> Sử dụng hàm "replace()" với tham số "regex=True" để thay thế các ký tự '-' trong cột "PulseRate" bằng 0.

- patientheartrate_2["PulseRate"].fillna('0', inplace =True): 

=> Sử dụng hàm "fillna()" kết hợp tham số "inplace=True" để điền trực tiếp giá trị 0 vào những ô trống trong cột "PulseRate".

```python
def avgPulserate(patientheartrate_2, firstname):
  sumPulserate = 0
  nsumPulserate = 0
  for i, value in enumerate(patientheartrate_2['Firstname']):
    if value == firstname:
      try:
        if patientheartrate_2['PulseRate'][i] !='0':
          sumPulserate += int(patientheartrate_2['PulseRate'].values[i])
          nsumPulserate += 1
      except:
        pass
  if nsumPulserate > 0:
    return str(sumPulserate/nsumPulserate)
  else:
    return 0
```

```python
def isValidIndex1(index, nrows):
  return index - 1 > 0 & index + 1 < nrows
```

```python
def isValidIndex2(index, nrows):
  return index + 1 < nrows & index + 2 < nrows
```

```python
def isPulseRateNotNull1(index, patientheartrate_2):
  return str(patientheartrate_2['PulseRate'].values[index-1])!= '0' and str(patientheartrate_2['PulseRate'].values[index+1])!= '0'
```

```python
def isPulseRateNotNull2(index, patientheartrate_2):
  return str(patientheartrate_2['PulseRate'].values[index+1])!= '0' and str(patientheartrate_2['PulseRate'].values[index +2])!= '0'
```

```python
def isInGroup1(patientheartrate_2, index, firstname):
  return patientheartrate_2['Firstname'].values[index-1]== firstname and patientheartrate_2['Firstname'].values[index +1]== firstname
```

```python
def isInGroup2(patientheartrate_2, index, firstname):
  return patientheartrate_2['Firstname'].values[index+1]== firstname and patientheartrate_2['Firstname'].values[index +2]== firstname
```

```python
def updatePulseRatena (index, patientheartrate_2):
  nrows = len(patientheartrate_2['PulseRate']) -2
  newvalue = ''
  firstname = patientheartrate_2['Firstname'].values[index]
  try:
    if isValidIndex1(index,nrows) & isPulseRateNotNull1(index, patientheartrate_2) & isInGroup1(patientheartrate_2, index, firstname):
      try: 
        newvalue = (int(patientheartrate_2['PulseRate'].values[index-1]) + int(patientheartrate_2['PulseRate'].values[index+1]))/2
      except:
        newvalue = '0'
    elif isValidIndex2(index,nrows) & isPulseRateNotNull2(index, patientheartrate_2) & isInGroup2(patientheartrate_2, index, firstname):
      try:
        newvalue = (int(patientheartrate_2['PulseRate'].values[index-1]) + int(patientheartrate_2['PulseRate'].values[index+1]))/2
      except:
        newvalue ='0'
    else:
      newvalue = avgPulserate(patientheartrate_2, firstname)
  except:
      newvalue = avgPulserate(patientheartrate_2, firstname)
  finally:
    pass
  return newvalue
```

```python
pulserateupdate = []
for index, value in enumerate(patientheartrate_2["PulseRate"]):
  if value =='0' or value == '':
    newvalue = updatePulseRatena(index,patientheartrate_2)
    pulserateupdate.append(newvalue)
  else:
    pulserateupdate.append(value)
```
