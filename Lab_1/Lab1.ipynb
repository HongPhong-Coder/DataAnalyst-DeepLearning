- Biến "i" trong vòng lặp "for" sẽ nhận lần lượt giá trị trong danh sách là 1,2,3.
- "print("TBM" + i)": Đây là phép nỗi chuỗi, tức kết quả trả về sẽ là giá trị của "i" sẽ được nối vào sau chuỗi "TBM". 

```python
for i in ["1", "2", "3"]:
    data["TBM" + i] = (data["T" + i]*2 + data["L" + i] + data["H" + i] + data["S" + i] + data["V" + i] * 2 + data["X" + i] + data["D" + i] + data["N" + i]) / 10
# data.rename(columns = {"TBM6":"TBM3"}, inplace = False)
data.head()
```

```python
# Tạo biến TBM1, TBM2, TBM3
data['TBM1'] = (data['T1']*2 + data['L1'] + data['H1'] + data['S1'] + data['V1']*2 + data['X1'] + data['D1'] + data['N1']) / 10
data['TBM2'] = (data['T2']*2 + data['L2'] + data['H2'] + data['S2'] + data['V2']*2 + data['X2'] + data['D2'] + data['N2']) / 10
data['TBM3'] = (data['T3']*2 + data['L3'] + data['H3'] + data['S3'] + data['V3']*2 + data['X3'] + data['D3'] + data['N3']) / 10
```

- Các cột "TBM1", "TBM2", "TBM3": là điểm trung bình môn cho các môn học 1,2,3. Kết quả sẽ được lưu vào các cột tương ứng trong DataFrame "data".

```python
def xep_loai(TBM):
    if TBM < 5.0:
        return 'Y'
    elif 5.0 <= TBM < 6.5:
        return 'TB'
    elif 6.5 <= TBM < 8.0:
        return 'K'
    elif 8.0 <= TBM < 9.0:
        return 'G'
    else:
        return 'XS'
    
data['XL1'] = data['TBM1'].apply(xep_loai)
data['XL2'] = data['TBM2'].apply(xep_loai)
data['XL3'] = data['TBM3'].apply(xep_loai)
```

- Hàm "apply": được sử dụng để tạo cột mới bằng những thông tin mới trong DataFrame có tên là "XL1", "XL2", "XL3".

```python
# Sử dụng Min-Max Normalization 
def min_max_normalization(series):
    # return (series - series.min()) / (series.max() - series.min()) * 4
    return (series * 4) / 10

data['US_TBM1'] = min_max_normalization(data['TBM1'])
data['US_TBM2'] = min_max_normalization(data['TBM2'])
data['US_TBM3'] = min_max_normalization(data['TBM3'])

data.head(10)
```

- Sử dụng phương pháp Min-Max Normalization lên các cột để chuẩn hóa dữ liệu về khoảng [0, 4].
- Áp dụng phương pháp đó lên các cột 'TBM1', 'TBM2', 'TBM3' rồi lưu kết quả vào các cột 'US_TBM1', 'US_TBM2', 'US_TBM3'.
- "data.head(10)": Hiển thị 10 dòng đầu tiên để xem kết qủa từ DataFrame.

```python
def ket_qua_xet_tuyen(KT, DH1, DH2, DH3):
    if KT in ['A', 'A1']:
        diem_trung_binh = (DH1*2 + DH2 + DH3) / 4
    elif KT == 'B':
        diem_trung_binh = (DH1 + DH2*2 + DH3) / 4
    else:
        diem_trung_binh = (DH1 + DH2 + DH3) / 3
    
    return 1 if diem_trung_binh >= 5.0 else 0

data['KQXT'] = data.apply(lambda row: ket_qua_xet_tuyen(row['KT'], row['DH1'], row['DH2'], row['DH3']), axis=1)

data.head(10)
```

- "data['KQXT']": Tạo 1 cột mới trong DataFrame "data" có tên là "KQXT" để lưu kết quả của từng sinh viên.
- "data.apply(lambda row: ket_qua_xet_tuyen(row['KT'], row['DH1'], row['DH2'], row['DH3']), axis=1)":
+, Hàm lamda nhận vào mỗi hàng của DataFrame và gọi hàm "ket_qua_xet_tuyen" với các đối số của từng hàng để tính kết quả xét tuyển của sinh viên. 
+, "axis=1": Áp dụng hàm lamda cho mỗi hàng trong DataFrame "data".
- "data.head(10)": Hiển thị 10 dòng đầu để xem kết quả của DataFrame.
